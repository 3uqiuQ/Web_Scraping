**Beautiful Soup - quotes.toscrape**

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'http://quotes.toscrape.com'
content = requests.get(url)
soup = BeautifulSoup(content.text, 'html.parser')
```

Mediante 'Inspeccionar', he averiguado que:  
- Las frases estan almacenadas en la página web en elementos <span> con la clase 'text'.  
- Los autores estan almacenados en elementos <small> con la clase 'author'.  
    
Eso es lo que le pido a BeautifulSoup que busque.

```python
quotes = soup.find_all('span', class_ ='text')

for quote in quotes:
    print(quote.get_text())
```

```python
authors = soup.find_all('small', class_ ='author')

for author in authors:
    print(author.get_text())
```

```python
frases = pd.DataFrame()

frases['Autores'] = authors
frases['Frases'] = quotes

pd.options.display.max_colwidth = None
frases
```

**Selenium - Bolsa de Madrid**

```python
import time
import pandas as pd
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from datetime import date
from datetime import datetime

# Configuramos las opciones de Selenium.
options = Options()
options.add_argument('--headless')
options.add_argument('--incognito')
driver = webdriver.Chrome(options=options)

# Definimos la URL a utilizar.
url_bolsa = 'https://www.bolsasymercados.es/bme-exchange/es/Indices/Resumen'

# Usamos selenium con las opciones definidas para navegar por la URL seleccionada.
driver.get(url_bolsa)
# Retrasamos el proceso 5 segundos para dar tiempo a que la página se cargue correctamente.
time.sleep(5)

# Permite acceder al HTML de la página.
html = driver.page_source

# Cerramos el WebDriver una vez finalizado
driver.quit()

# Analiza el HTML usando BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')

# Encuentra y muestra la tabla.
results = soup.find('table', {'role': 'table', 'class': ''})
if results:
    print(results.prettify())
else:
    print("No se encontró el elemento deseado en la página.")

now = datetime.now()  
```

```python
tabla_indices = soup.find_all('td', {'class':'text-nowrap', 'role':'rowheader'})
tabla_indices
```

```python
indices = []
valores = []
```

```python
for td in tabla_indices:
    # Busca la etiqueta 'a' dentro de cada elemento 'td'.
    url_tag_ind = td.find('a')
    # Verifica si se encontró una etiqueta 'a' dentro del 'td'.
    if url_tag_ind:
        # Accede al texto de la etiqueta de la URL.
        tag_ind = url_tag_ind.text
        # Imprime la etiqueta.
        print(tag_ind)
        indices.append(tag_ind)
    else:
        print('No se han encontrado')
```

```python
print(len(indices))
print(indices)
```

```python
indices_rv = indices[:85]
print(len(indices_rv))
```

```python
tabla_valores = soup.find_all('td', {'class':'text-right', 'role':'cell'})
print(len(tabla_valores))
tabla_valores
```

```python
for td in tabla_valores:
    if 'text-right' in td['class'] and 'cell' in td['role']:
        tag_v = td.text
        print(tag_v)
        valores.append(tag_v)
```

```python
# Iteramos en los valores cada 5, para obtener los de la primera columna.

valores2 = []

j = 0
while j < len(valores):
    valores2.append(valores[j])
    j += 5
```

```python
valores_rv = valores2[:85]
print(len(valores_rv))
```

```python
# Mostramos los índices y los valores de Renta Variable.

tabla_bolsa = pd.DataFrame()
tabla_bolsa['Indices'] = indices_rv
tabla_bolsa['Valores'] = valores_rv

print(now)
tabla_bolsa
```

**Exercici 3**

Tria una pàgina web que tu vulguis i realitza web scraping mitjançant la llibreria Selenium primer i Scrapy després.

He elegido una página web de ciclismo, donde se recopilan vídeos e información de los puertos de montaña más importantes.  
He copiado una tabla-resumen de los mismos.

```python
import time
import pandas as pd
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from datetime import date
from datetime import datetime


# Configuramos las opciones de Selenium.
options = Options()
options.add_argument('--headless')
options.add_argument('--incognito')
driver = webdriver.Chrome(options=options)

# Definimos la URL a utilizar.
url_col_collective = 'http://thecolcollective.com/col-collection/all'

# Usamos selenium con las opciones definidas para navegar por la URL seleccionada.
driver.get(url_col_collective)
# Retrasamos el proceso 5 segundos para dar tiempo a que la página se cargue correctamente.
time.sleep(5)

# Permite acceder al HTML de la página.
html = driver.page_source

# Cerramos el WebDriver una vez finalizado
driver.quit()

# Analiza el HTML usando BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')

# Encuentra y muestra la tabla.
results = soup.find('div', class_= 'stats-table')
if results:
    print(results.prettify())
else:
    print("No se encontró el elemento deseado en la página.")
    
now_2 = datetime.now()
```

```python
nom_puertos = soup.find_all('div', class_='col-md-2 col-sm-3 col-xs-6 col-left')
nombres_puertos = []
for puerto in nom_puertos:
    nombres_puertos.append(puerto.text)
print(nombres_puertos)
```

```python
nom_poblaciones = soup.find_all('div', class_='col-md-2 col-sm-3 hidden-xs col-left')
nombres_poblaciones = []
for poblacion in nom_poblaciones:
    nombres_poblaciones.append(poblacion.text)
print(nombres_poblaciones)
```

```python
longitudes = soup.find_all('div', class_ = 'col-md-1 col-sm-1 col-xs-3')
long_puertos = []
for longitud in longitudes:
    long_puertos.append(longitud.text)
print(long_puertos)
```

```python
alturas = soup.find_all('div', class_='col-md-1 hidden-sm hidden-xs')
desnivel_box = []
desnivel = []
for altura in alturas:
    desnivel_box.append(altura.text)
    
d = 0
while d < len(desnivel_box):
    desnivel.append(desnivel_box[d])
    d += 3
    
print(desnivel)
```

```python
pendientes = soup.find_all('div', class_ = 'col-md-1 col-sm-2 col-xs-3')
porcentaje = []
for pendiente in pendientes:
    porcentaje.append(pendiente.text)
print(porcentaje)
```

```python
tabla_puertos = pd.DataFrame()
tabla_puertos['Puerto'] = nom_puertos
tabla_puertos['Poblacion'] = nombres_poblaciones
tabla_puertos['Longitud'] = long_puertos
tabla_puertos['Desnivel'] = desnivel
tabla_puertos['Porcentaje'] = porcentaje

pd.set_option('display.max_rows', None)     # Mostrar todas las filas sin límite

tabla_puertos = tabla_puertos.iloc[list(range(1, 49)) + list(range(51, 67))]

print(now_2)
tabla_puertos
```
